```python
import pandas as pd

data = {
    "Name": ["Ali", "osman", "haider","imad","hanan","Sara", "Omar"],
    "Age": [25, 30,40,50,10,20, 22],
    "City": ["Lahore","multan","sargodha","sahiwal","naran", "Karachi", "Islamabad"]
}

df = pd.DataFrame(data)
print(df)
```

```python
import numpy as np
arr = np.array([1, 2, 3, 4])#for single numpay array
print(arr)
```

```python
matrix = np.array([[1, 2], [3, 4]]) # multi dimensional array
print(matrix)
```

```python
arr = np.arange(1, 10, 2)   # 1 se 9 tak,
print(arr)
```

```python
print(np.zeros((2,3)))   # 2 row 3 column, sab 0
print(np.ones((3,2)))    # 3 row 2 column, sab 1
```

```python
arr = np.array([1,2,3])# operators
print(arr + 5)     # her element me 5 add hoga
print(arr * 2)     # her element 2 se multiply
```

```python
arr = np.array([10, 20, 30, 40])
print(arr.mean())    # Average
print(arr.max())     # Maximum value
print(arr.min())     # Minimum value
```

```python
arr = np.arange(1, 9)
print(arr.reshape(2, 4))   # 1D array ko 2by4 matrix bnana
```

```python
arr = np.random.random((2,3)) #random values
print(arr)
```

```python
arr = np.random.randint(10,100,(3,4))
print(arr)
```

```python
arr = np.linspace(10,30,6)
print(arr)
```

bar plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['english','urdu','french','japnese','hindi']
people = [100,200,300,400,500]
ax.bar(languages,people)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['english','urdu','french','japnese','hindi']
people = [100,200,300,400,500]
ax.bar(languages,people)
ax.set_xlabel('languages')
ax.set_ylabel('number of people')
ax.set_title('bar plot')
plt.show()
```

pie chart


```python
fig1 = plt.figure()
ax1 = fig1.add_axes([0,0,1,1])
languages = ['english','urdu','french','japnese','hindi']
people = [100,200,300,400,500]
ax1.pie(people,labels=languages, autopct = '%1.1f%%')
plt.show()
```

```python
x = [5, 7, 8, 7, 6]
y = [99, 86, 87, 88, 100]

plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax2 = fig2.add_axes([0,0,1,1])
ax2.scatter(x,y, color='g')
ax2.scatter(x,z, color='b')
plt.show()
```
